```python
import os, shutil, random

DATASET_PATH = "/content/dataset/cancer data sets 9 classes"
train_dir = "/content/train"
test_dir  = "/content/test"

classes = [d for d in os.listdir(DATASET_PATH) if os.path.isdir(os.path.join(DATASET_PATH, d))]

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

for cls in classes:
    class_path = os.path.join(DATASET_PATH, cls)

    # Recursively find all image files in this class folder
    imgs = []
    for root, _, files in os.walk(class_path):
        for f in files:
            if f.lower().endswith((".jpg", ".jpeg", ".png")):
                imgs.append(os.path.join(root, f))

    if not imgs:
        print(f"No images found for class {cls}!")
        continue

    os.makedirs(os.path.join(train_dir, cls), exist_ok=True)
    os.makedirs(os.path.join(test_dir, cls), exist_ok=True)

    random.shuffle(imgs)
    split = int(0.8 * len(imgs))
    train_imgs = imgs[:split]
    test_imgs  = imgs[split:]

    # Copy train images
    for img in train_imgs:
        shutil.copy(img, os.path.join(train_dir, cls, os.path.basename(img)))

    # Copy test images
    for img in test_imgs:
        shutil.copy(img, os.path.join(test_dir, cls, os.path.basename(img)))

print("✅ Train/Test split complete.")
```

```python
import os, shutil

# Paths
train_dir = "/content/train"
test_dir  = "/content/test"

# Remove the unwanted folder
for folder in ["cancer data sets 9 classes"]:
    train_path = os.path.join(train_dir, folder)
    test_path  = os.path.join(test_dir, folder)

    if os.path.exists(train_path):
        shutil.rmtree(train_path)
        print(f"Removed {train_path}")

    if os.path.exists(test_path):
        shutil.rmtree(test_path)
        print(f"Removed {test_path}")
```

```python
import torch
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Image preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),      # ViT expects 224x224
    transforms.ToTensor(),
    transforms.Normalize([0.5, 0.5, 0.5], [0.5, 0.5, 0.5])
])

# Dataset objects
train_data = datasets.ImageFolder(train_dir, transform=transform)
test_data  = datasets.ImageFolder(test_dir, transform=transform)

# DataLoaders
train_loader = DataLoader(train_data, batch_size=16, shuffle=True)
test_loader  = DataLoader(test_data, batch_size=16)

# Class info
class_names = train_data.classes
num_classes = len(class_names)

print("Classes:", class_names)
print("Number of classes:", num_classes)
print("Number of train images:", len(train_data))
print("Number of test images:", len(test_data))
```

```python
from transformers import ViTForImageClassification

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", DEVICE)

# Load pre-trained ViT model with correct number of classes
model = ViTForImageClassification.from_pretrained(
    "google/vit-base-patch16-224-in21k",
    num_labels=num_classes
).to(DEVICE)

print("ViT model loaded successfully!")
```

```python
# Optimizer & loss function
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
loss_fn = torch.nn.CrossEntropyLoss()

# Number of epochs
EPOCHS = 3 # You can increase if GPU allows
```

```python
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0

    for images, labels in train_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        optimizer.zero_grad()
        outputs = model(pixel_values=images).logits  # ViT expects pixel_values
        loss = loss_fn(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {total_loss/len(train_loader):.4f}")
```

```python
model.eval()
correct, total = 0, 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        outputs = model(pixel_values=images).logits
        preds = torch.argmax(outputs, dim=1)

        correct += (preds == labels).sum().item()
        total += labels.size(0)

accuracy = correct / total * 100
print(f"✅ Test Accuracy: {accuracy:.2f}%")
```

```python
import numpy as np

model.eval()
all_preds = []
all_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        outputs = model(pixel_values=images).logits
        preds = torch.argmax(outputs, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy  = accuracy_score(all_labels, all_preds)
precision = precision_score(all_labels, all_preds, average="weighted")
recall    = recall_score(all_labels, all_preds, average="weighted")
f1        = f1_score(all_labels, all_preds, average="weighted")

print(f"✅ Accuracy : {accuracy*100:.2f}%")
print(f"✅ Precision: {precision*100:.2f}%")
print(f"✅ Recall   : {recall*100:.2f}%")
print(f"✅ F1-Score : {f1*100:.2f}%")
```

```python
from sklearn.metrics import classification_report

print("\n📊 Classification Report:\n")
print(classification_report(all_labels, all_preds, target_names=class_names))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(all_labels, all_preds)
print("Confusion Matrix:\n")
print(cm)
```

```python
import os, shutil, random

SOURCE_TEST_DIR = "/content/test"
BALANCED_TEST_DIR = "/content/balanced_test"
IMAGES_PER_CLASS = 100   # you can change (800, 1000, etc.)

os.makedirs(BALANCED_TEST_DIR, exist_ok=True)

for cls in os.listdir(SOURCE_TEST_DIR):
    src_cls = os.path.join(SOURCE_TEST_DIR, cls)
    dst_cls = os.path.join(BALANCED_TEST_DIR, cls)
    os.makedirs(dst_cls, exist_ok=True)

    images = [f for f in os.listdir(src_cls)
              if f.lower().endswith((".jpg",".png",".jpeg"))]

    if len(images) < IMAGES_PER_CLASS:
        raise ValueError(f"{cls} has only {len(images)} images!")

    selected = random.sample(images, IMAGES_PER_CLASS)

    for img in selected:
        shutil.copy(os.path.join(src_cls, img),
                    os.path.join(dst_cls, img))

print("✅ Balanced test set created")
```

```python
for cls in os.listdir(BALANCED_TEST_DIR):
    print(cls, len(os.listdir(os.path.join(BALANCED_TEST_DIR, cls))))
```

```python
balanced_test_data = datasets.ImageFolder(
    BALANCED_TEST_DIR,
    transform=transform
)

balanced_test_loader = DataLoader(
    balanced_test_data,
    batch_size=16,
    shuffle=False
)

print("Total balanced test images:", len(balanced_test_data))
```

```python
from sklearn.metrics import confusion_matrix

y_true, y_pred = [], []

model.eval()
with torch.no_grad():
    for images, labels in balanced_test_loader:
        images = images.to(DEVICE)
        labels = labels.to(DEVICE)

        outputs = model(pixel_values=images).logits
        preds = torch.argmax(outputs, dim=1)

        y_true.extend(labels.cpu().numpy())
        y_pred.extend(preds.cpu().numpy())

cm = confusion_matrix(y_true, y_pred)
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(cm,
            annot=True,
            fmt="d",
            cmap="Blues",
            xticklabels=class_names,
            yticklabels=class_names)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix (Balanced Test Set)")
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

acc  = accuracy_score(y_true, y_pred)
prec = precision_score(y_true, y_pred, average="macro")
rec  = recall_score(y_true, y_pred, average="macro")
f1   = f1_score(y_true, y_pred, average="macro")

print(f"Accuracy  : {acc:.4f}")
print(f"Precision : {prec:.4f}")
print(f"Recall    : {rec:.4f}")
print(f"F1-score  : {f1:.4f}")
```
